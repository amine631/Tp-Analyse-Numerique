```python
import matplotlib . pyplot as plt
import numpy as np

```

```python
def f ( x ):
    return x **3+ x**2- 3*x - 3
x=np.linspace ( -2 ,2)
y=f(x)
plt.title("solution dans [-2,2]");
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.plot (x,y)
plt.axhline()
plt.show()
```

```python
def dichotomie (f ,a ,b , epsilon ):

    g , d = min (a , b ) , max (a , b )
    valg , vald = f ( g ) , f ( d )
    n = 0
    while d - g > epsilon :
        n += 1
        m = ( g + d )/2
        valm = f ( m )
        if valg * valm < 0 :
            d = m
            vald = valm
        else :
            g = m
            valg = valm
    print (" nombre d'itérations nécessaires :", n )
    return ( g + d )/2
```

```python
dichotomie ( f, -2 , 2 , 0.001)
```

```python
x3=dichotomie(f,-2,-1.5,10)
print(x3)
```

```python
dichotomie ( f, -2 , -1.5 , 0.0001)
```

```python
intervalle=[[-2,-1.5],[-1.5,-0.5],[1.5,2]]
for k in intervalle:
    print(dichotomie(f,k[0],k[1],0.001))
```

```python
dichotomie ( f, -2 , -1.5 , 0.00001)
```
